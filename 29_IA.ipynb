```python
# prompt: Escribe una funcion en Python que calcule el area de un circulo dado su radio

import math

def area_circulo(radio):
  """Calcula el área de un círculo dado su radio.

  Args:
    radio: El radio del círculo.

  Returns:
    El área del círculo.
  """
  area = math.pi * radio**2
  return area
```

```python
area_circulo(15)
```

```python
# prompt: Escribe un codigo en Python que genere un grafico de barras usando matplotlib y los datos para eje x y eje y seran obtenidos al azar y seran 15 datos

import matplotlib.pyplot as plt
import random

# Generar datos aleatorios para el eje x e y
x = list(range(1, 16))  # Eje x: 15 datos
y = [random.randint(1, 50) for _ in range(15)]  # Eje y: 15 datos aleatorios entre 1 y 50

# Crear el gráfico de barras
plt.bar(x, y)

# Agregar etiquetas a los ejes
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Agregar un título al gráfico
plt.title("Gráfico de Barras")

# Mostrar el gráfico
plt.show()
```

```python
# prompt: usa los datos anteriores para crear un archivo csv, graba dicho archivo, leelo y grafica los datos usando Pandas y Matplotlib, muestra al final una tabla con los datos

import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame de Pandas con los datos
data = {'Eje X': x, 'Eje Y': y}
df = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV
df.to_csv('datos.csv', index=False)

# Leer el archivo CSV usando Pandas
df = pd.read_csv('datos.csv')

# Graficar los datos usando Matplotlib y Pandas
plt.figure(figsize=(10, 6))
plt.bar(df['Eje X'], df['Eje Y'])
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Barras desde CSV')
plt.show()

# Mostrar la tabla con los datos
print(df)
```

```python
# prompt: genera la grafica sobre dataframe anterior, dandole un tamaño menor y cambiando el color de las columnas y generando una linea en otro color que sea el resultado de la media (linea sobre media de datos)

import matplotlib.pyplot as plt

# Calcular la media de los datos en el eje Y
media_y = df['Eje Y'].mean()

# Graficar los datos usando Matplotlib y Pandas
plt.figure(figsize=(6, 4))  # Tamaño de la figura más pequeño
plt.bar(df['Eje X'], df['Eje Y'], color='skyblue')  # Cambiar el color de las barras
plt.axhline(y=media_y, color='red', linestyle='-')  # Línea horizontal para la media en rojo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Barras desde CSV')
plt.show()
```
