#Grafica de caja

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Datos de ejemplo
datos = {
    "Categoria": ["A", "A", "B", "B", "A", "B", "A", "B", "A", "B"],
    "Valor": [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
}
```

```python
df = pd.DataFrame(datos)
df.head()
```

```python
# Creamos grafica de caja
sns.boxplot(x="Categoria", y="Valor", data=df)
plt.title("Grafica de Caja")
plt.show()
```

#Grafica de correlacion

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Datos de ejemplo
datos = np.random.rand(10, 12)
df = pd.DataFrame(datos, columns=[f"Var{i}" for i in range(12)])
df.head()
```

```python
# Matriz de correlacion
correlacion = df.corr()
```

```python
# Grafica de mapa de calor
sns.heatmap(correlacion, annot=True, cmap="coolwarm")
plt.title("Grafica de Correlacion")
plt.show()
```

#Personalizacion

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Datos de ejemplo
datos = {
    "Categoria": ["A", "A", "B", "B", "A", "B", "A", "B", "A", "B"],
    "Valor": [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
}
```

```python
df = pd.DataFrame(datos)
df.head()
```

```python
# Creamos grafica, estilo y color
sns.set(style="whitegrid", palette="pastel")
sns.boxplot(x="Categoria", y="Valor", data=df)
plt.title("Grafica Personalizada")
plt.show()
```

#Ejercicio practico

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# carga de datos
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Creamos grafica de caja
plt.figure(figsize=(10, 6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Grafica de Caja de Propinas por dia")
plt.show()
```

```python
# Creamos grafica de violin
plt.figure(figsize=(10,6))
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Grafica de Violin de Propinas por dia")
plt.show()
```

```python
# Creamos mapa de calor
correlacion = tips.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(correlacion, annot=True, cmap="coolwarm")
plt.title("Grafica de Correlacion de Propinas")
plt.show()
```
